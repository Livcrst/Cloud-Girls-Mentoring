Biblotecas Python: Conjunto de funções e módulos úteis para o usuário, para usa-las precisamos instalar ou importar em nosso ambiente. 
Para que isso seja possivel precisamos usar o pip que é nosso gerenciador de pacotes e bibliotecas Python.

```python
#Exemplo de instalação, matplotlib pelo pip
# O simnpolo a frente de pip install serve  
%pip install matplotlib
```

```python
#instalar uma versão especifica da lib, com esse comando o pip procura a versão que indicamos, exclui a versão anterior instalada e instala a versão que indicamos.

%pip install matplotlib==3.6.2
```

Importando e usando as libs importadas

```python
estudantes = ["joao","Maria", "Pedro", "Joana", "Ana"]
notas = [8,7,9,5,6]
```

```python
# Construindo grafico de barras

plt.bar(estudantes, notas)
plt.show()
```

```python
# Usando a biblioteca random 

from random import choice #

# help(choice) #Essa função explica como podemos usar uma determinada função

print(choice(estudantes))
```
